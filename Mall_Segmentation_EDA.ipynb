```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head(10).plot.bar(x = 'Age', y = 'Annual Income (k$)' )
plt.title('Age v/s Annual Income')
```

```python
df.head(10).plot.bar(x = 'Age', y = 'Spending Score (1-100)')
plt.title('Age v/s Spending Score')
```

```python
df.sort_values( by = ['Annual Income (k$)'], ascending = False).head(10).plot.bar(x = 'Age', y = 'Annual Income (k$)')
```

```python
df.columns
```

```python
subset = df[df.Gender.isin(['Male', 'Female'])]
sns.set_style('darkgrid')
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
ax = sns.violinplot(x = 'Age', y = 'Gender', data = subset)
ax.set_title('Violinplot')
```

```python
subset = df[df['Spending Score (1-100)']>60]
p = sns.jointplot(x = 'Gender', y = 'Spending Score (1-100)', data = subset)
```

```python
plt.rcParams['figure.figsize'] = (3,3)
sm = sns.heatmap(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].corr(), annot = True, cmap= 'Blues')
sm.set_title('Heatmap Distribution', fontsize = 16)
```

```python
df.sort_values( by = ['Spending Score (1-100)'], ascending = False).head(5).plot.bar(x = 'Age', y = 'Spending Score (1-100)')
```

```python
subset = df[df['Spending Score (1-100)']>65]
p = sns.violinplot(x = 'Gender', y = 'Spending Score (1-100)', data = subset)
```
